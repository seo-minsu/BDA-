###2) 이상치 확인 및 처리
(상자그림 활용, 1.5*IQR 초과하는 값 = 이상치)

```
## 타이타닉 데이터에서 상자그림

sns.boxplot(df['age'])
```

```
##Q1, Q3, IQR 구하기

Q1 = df['age'].quantile(.25)
Q3 = df['age'].quantile(.75)
IQR = Q3-Q1

print(Q1, Q3, IQR)
```

```
upper = Q3 + 1.5*IQR
lower = Q1 - 1.5*IQR
print(upper,',', lower)
```

```
## 문제: age 변수의 이상치를 제외한 데이터 수는? (상자그림 기준)
cond1 = (df['age']<=upper)  ##부호주의
cond2 = (df['age']>=lower)

print(len(df[cond1&cond2]))
print(len(df[cond1]))
print(len(df))
```

##7. 날짜/시간 데이터, index 다루기

###1) 날짜 다루기

```
#데이터 만들기

df = pd.DataFrame({
    '날짜': ['20230105', '20230105', '20230223', '20230223', '20230312', '20230422', '20230511'],
    '물품': ['A','B','A','B','A','B','A'],
    '판매수': [5, 10, 15, 15, 20, 25, 40],
    '개당수익': [500, 600, 500, 600, 600, 700, 800]})

df
```

```
df.info()
```

```
##데이터 타입 datetime으로 변경

df['날짜'] = pd.to_datetime(df['날짜'])
df.info()
```

```
## 년, 월, 일 변수(열) 추가

df['year'] = df['날짜'].dt.year
df['month'] = df['날짜'].dt.month
df['day'] = df['날짜'].dt.day
df
```

```
##날짜 구간 필터링

df[df['날짜'].between('2023-01-01', '2023-02-23')] ##좌우 모두 포함

##(주의) 날짜와 시간이 같이 있는 데이터에 between 함수를 쓸 경우 형식이 동일해야 함.
## (ex) 2023-01-05 12:30:05 → between('2023-01-05 12:00:00', '2023-01-05 12:40:00' ) 이렇게 통일시켜줘야 함.
```

```
## 날짜를 인덱스로 설정 후 loc 함수 사용

## 데이터 만들기

df = pd.DataFrame({
    '날짜': ['20230105', '20230105', '20230223', '20230223', '20230312', '20230422', '20230511'],
    '물품': ['A','B','A','B','A','B','A'],
    '판매수': [5, 10, 15, 15, 20, 25, 40],
    '개당수익': [500, 600, 500, 600, 600, 700, 800]})

df['날짜'] = pd.to_datetime(df['날짜'])

df = df.set_index('날짜')  ## drop = True(디폴트) or False

df.head(3)
```

```
##loc 함수는 날짜가 index되어있는 상태에서 쓸 수 있음

print(df.loc['2023-01-05':'2023-02-23']) ##둘다 기간 포함

print(df.loc[(df.index >= '2023-01-05') & (df.index <= '2023-02-23') ])
```
