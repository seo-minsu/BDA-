###2. 기초통계량(평균, 중앙값, IQR, 표준편차 등)

```
df = pd.read_csv('mtcars.csv')
df.head()
```

```
## 중심측도를 나타내는 값(평균, 중앙값, 최빈값)
## 32: 행, 12: 열(변수)
df.shape
```

```
## 평균
mpg_mean = df['mpg'].mean()
print(mpg_mean)
```

```
## 중앙값
mpg_median = df['mpg'].median()
print(mpg_median)
```

```
## 최빈값
mpg_mode = df['cyl'].mode()
print(mpg_mode)
```

```
print(mpg_mode[0])
```

```
df['cyl'].value_counts()
```

```
## 산포도를 나타내는 값(분산, 표준편차, IQR, 범위(최대-최소) 등)
```

```
# 분산
mpg_var=df['mpg'].var()
print(mpg_var)
```

```
# 표준편차
mpg_std = df['mpg'].std()
print(mpg_std)
```

```
# IQR
Q1 = df['mpg'].quantile(0.25)
print(Q1)
```

```
Q3 = df['mpg'].quantile(0.75)
print(Q3)
```

```
IQR = Q3-Q1
print(IQR)
```

```
Q2 = df['mpg'].quantile(0.5)
print(Q2)

print(df['mpg'].median())
```

```
## 범위

mpg_max = df['mpg'].max()
mpg_min = df['mpg'].min()
print(mpg_max)
print(mpg_min)

mpg_range = mpg_max-mpg_min
print(mpg_range)
```

##3) 분포의 비대칭도

```
#왜도

mpg_skew = df['mpg'].skew()
print(mpg_skew)

#첨도
mpg_kurt = df['mpg'].kurt()
print(mpg_kurt)
```

##3) 데이터 필터링

```
## 2개 조건 필터링 후 데이터 개수(or)
print(len(df[cond1 | cond2]))
```

##5) 데이터 변경(조건문)

```
import numpy as np

df = pd.read_csv('mtcars.csv')

## np.where 함수 활용
## hp 변수 값 중에서 205가 넘는 값은 205로 처리, 나머지는 그대로 유지하기
df['hp'] = np.where(df['hp']>=205, 205, df['hp'])

## 내림차순 정렬
df.sort_values('hp', ascending = False).head(10)
```
