```python
## 문제 11
## cyl 가 4인 자동차와 6인 자동차 그룹의 mpg 평균값 차이를 절대값 정수로 출력

df = pd.read_csv('mtcars.csv')

cond1 = (df['cyl']==4)
cond2 = (df['cyl']==6)
cyl4 = df[cond1]['mpg'].mean()
cyl6 = df[cond2]['mpg'].mean()

print(round(abs(cyl4-cyl6)))
```

```python
##문제 12

## hp 변수에 대해 데이터표준화(z-score)를 진행하고 이상치의 수 구하라. (단, 이상치는 Z값이 1.5를 초과하거나 -1.5 미만인 값임)

## Z = (X-평균) / 표준편차
std_hp = df['hp'].std()
mean_hp = df['hp'].mean()
df['zscore'] = (df['hp']-mean_hp) / std_hp
df
cond1 = (df['zscore']>1.5)
cond2 = (df['zscore']<-1.5)
print(len(df[cond1])+len(df[cond2]))
```

```python
##문제13

## mpg 칼럼을 최대최소 scaling 하고, 0.7보다 큰 값을 가지는 레코드 수 구하라.

## min-max scaling  = (X-min) / (max-min)

min_mpg = df['mpg'].min()
max_mpg = df['mpg'].max()
df['mpg'] = (df['mpg']-min_mpg) / (max_mpg-min_mpg)

cond1 = (df['mpg']>0.7)
print(len(df[cond1]))
```

```python
##문제 14

## wt칼럼에서 상자그림 기준으로 이상치의 개수를 구하라.

Q1 = df['wt'].quantile(0.25)
Q3 = df['wt'].quantile(0.75)

IQR = Q3-Q1

upper = Q3 + 1.5*IQR
lower = Q1 - 1.5*IQR

cond1 = (df['wt']>upper)
cond2 = (df['wt']<lower)

print(len(df[cond1])+len(df[cond2]))
```

```python
##문제 15

## 판매수 컬럼의 결측치를 최소값으로 대체하고
## 결측치가 있을 때와 최소값으로 대체했을 때 평균값의 차이를 절대값 정수로 출력하라.

df = pd.DataFrame( {
    '날짜': ['20220103','20220105', None,'20230127','20220203', '20220205','20230210', '20230223', '20230312', '20230422', '20220505', '20230511'],
    '제품' : ['A','B', None, 'B', 'A', None,'A', 'B', 'A', 'B', 'A', 'A'],
    '판매수': [3, None, 5, 10, 10, 10, 15, 15, 20, None, 30, 40],
    '개당수익': [300, 400, 500, 600, 400, 500, 500, 600, 600, 700, 600, 600] })

mean1 = df['판매수'].mean()

df2 = df.copy()
최소값 = df2['판매수'].min()
df2['판매수'] = df2['판매수'].fillna(최소값)

mean2 = df2['판매수'].mean()

print(round(abs(mean1-mean2)))
```

```python
##문제 16

## vs변수가 0이 아닌 차량 중에 mpg 값이 가장 큰 차량의 hp 값을 출력하라.

## vs 변수가 0이 아닌 차량 --> mpg 값 내림차순 뽑기 --> 그것의 hp 변수의 iloc의 0번째 값

## !=x : x가 아닌 것
## iloc[n] : n번째 인덱스 변수 값 추출

df = pd.read_csv('mtcars.csv')

df = df[df['vs']!=0]

df = df.sort_values('mpg', ascending=False)

#print(df)
print(df['hp'].iloc[0])

##loc 쓸거면 mpg 값이 가장 큰 값의 인덱스 번호를 직접 써야 함(19로 써야 65가 나옴)
```

```python
##문제 17

## gear 변수값이 3, 4인 두 그룹의 hp 표준편차값의 차이를 절대값 소수점 첫째자리까지 출력

## 2개 그룹으로 필터링 --> 표준편차
df = pd.read_csv('mtcars.csv')
gear3 = df[df['gear']==3]
gear4 = df[df['gear']==4]

std3 = gear3['hp'].std()
std4 = gear4['hp'].std()

print(round(abs(std3-std4), 1))
```

```python
##문제 18

## gear 변수의 값별로 그룹화하여 mpg 평균값 산출하고
## 평균값이 높은 그룹의 mpg 제3사분위수 값 구하라.

## gear, mpg 변수만 필터링 ==> gear 변수로 그룹화하여 mpg 평균값 보기 --> 평균값 높은 그룹의 mpg Q3 구하기

df = df.loc[:, ['gear', 'mpg']]

#print( df.groupby('gear').mean()) ## gear 4가 가장 높음.

gear4 = df[df['gear']==4]
print(gear4['mpg'].quantile(.75))
```

```python
##문제 19

## hp 항목의 상위 7번째 값으로 상위 7개 값을 변환한 후,
## hp가 150 이상인 데이터를 추출하여 hp의 평균값을 반올림하여 정수로 출력하라.

df = pd.read_csv('mtcars.csv')

## hp 열 기준 내림차순 ==> 인덱스 초기화 ==> hp 상위 7번째 값 확인  ==> np.whwere 활용

df = df.sort_values('hp', ascending=False)

df = df.reset_index(drop=True) ## drop=True는 기존 인덱스 삭제
#print(df)

top7 = df['hp'].loc[6]

import numpy as np
df['hp'] = np.where(df['hp']>=205, top7, df['hp'])

##np.where(조건, 조건에 해당할 값, 그렇지 않을 때의 값)

cond1 = (df['hp']>=150)
df = df[cond1]
print(round(df['hp'].mean()))
```

```python
##문제25

##2023년 9월 24일 16:00~22:00 사이에 전체 제품의 판매수를 구하라.

df = pd.DataFrame( {
    '물품' : ['A','B', 'A', 'B', 'A', 'B','A'],
    '판매수': [5, 10, 15, 15, 20, 25, 40],
    '개당수익': [500, 600, 500, 600, 600, 700, 600] })

time = pd.date_range('2023-09-24 12:25:00', '2023-09-25 14:45:30', periods = 7)

df['time'] = time
df = df[ ['time', '물품', '판매수', '개당수익']]
#-------------------------------------------------------------------------------
## 풀이

## 시간데이터 만들기
## 컬럼과 인덱스에 둘다 time 변수를 놓고 풀이
## 데이터 타입 datetime으로 변경(항상 df.info()로 데이터 타입 확인할 것)

df['time'] = pd.to_datetime(df['time'])

df = df.set_index('time', drop = False) ##drop=true 는 변수에서 사라지고 인덱스로만 설정, drop=false는 변수도 있고 인덱스로도 설정

# print(df)

df_after = df[df['time'].between('2023-09-24 00:00:00', '2023-09-24 23:59:59')]

## 시간 필터링 함수인 between time 쓰려면 time이 인덱스가 있어야 함.
##지금은 인덱스랑 변수에 둘 다 있으니까 아래처럼 필터링 해도 됨.

df = df[df['time'].between('2023-09-24 16:00:00', '2023-09-24 22:00:00')]

print(df['판매수'].sum())



```

```python
##문제 26

## 9월 25일 00:00~12:00까지의 B물품의 매출액 총합을 구하라.

df = pd.DataFrame( {
    '물품' : ['A','B', 'A', 'B', 'A', 'B','A'],
    '판매수': [5, 10, 15, 15, 20, 25, 40],
    '개당수익': [500, 600, 500, 600, 600, 700, 600] })

time = pd.date_range('2023-09-24 12:25:00', '2023-09-25 14:45:30', periods = 7)

df['time'] = time
df = df[ ['time', '물품', '판매수', '개당수익']]

df = df.set_index('time', drop=True)

#------------------------------------------------------------------------------

## 날짜랑 시간 필터링, 매출액 파생변수 생성, 합

df['매출액'] = df['판매수']*df['개당수익']

df = df.reset_index()
## reset 쓴 이유는 time 변수 사용하려고, index 초기화하기 위해서
## default = drop Flase 임

df['time'] = pd.to_datetime(df['time'])

df = df[df['time'].between('2023-09-25 00:00:00', '2023-09-25 12:00:00')]

df = df[df['물품']=='B']

print(df['매출액'].sum())
```
